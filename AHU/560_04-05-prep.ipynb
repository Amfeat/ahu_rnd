```python
flow_columns = []
for name in  df.columns:
    if name.endswith('Flow'):
        flow_columns.append(name)
# flow_columns
```

```python
temp_columns = []
for name in  full_df.columns:
    if name.endswith('SpaceTemp'):
        temp_columns.append(name)
# temp_columns
```

```python
df = full_df[actual_columns]
df.loc[:, 'Total_flow'] = full_df[flow_columns].sum(axis=1)
# df['Total_flow'].plot()
```

```python
df = df.copy()
df.loc[:,'temp_delta'] = full_df["ReturnTemp.AC3"] - full_df["SupplyTemp.AC3"]
df.loc[:,'Q_load'] = df['temp_delta'] * df['Total_flow']
# df[['temp_delta', 'Q_load', 'Total_flow', 'OAT']].plot()
```

```python
df.loc[:,'median_space_temp'] = full_df[temp_columns].median(axis=1)
```

```python
df = df.copy()
df.loc[:,'resampled'] = df['median_space_temp']
df.loc[:,'resampled'] = df['resampled'].resample('100T').interpolate(method='spline', order=5)

df.loc[:,'resampled'] = df.resampled.interpolate(method='time', order = 1)

# df[['resampled', 'median_space_temp']].plot()
```

```python
import numpy as np
tmp_ = df['resampled']
slope = pd.Series(np.gradient(tmp_.values), tmp_.index, name='slope')


df.loc[:,'der'] = slope
# df['der'].plot()
```
